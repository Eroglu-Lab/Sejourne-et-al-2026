```python
#Made by Sarah J for Triple Nuclear Analysis

import os
import tkinter as tk
from tkinter import filedialog
from pathlib import Path
get_ipython().run_line_magic("gui", "tk")
root = tk.Tk()
root.withdraw()
root.call("wm", "attributes", ".", "-topmost", True)


tif_path = Path(filedialog.askdirectory(title="Select your folder titled 'Restored'"))
                            
folder_path = Path(filedialog.askdirectory(title="Select your folder titled 'Tile Scans' that contains all of the individual tile scan folders"))

folder_list = os.listdir(folder_path) #list of the folders in the Tile Scans folder
tif_list = os.listdir(tif_path) #list of the .tif images to move

for folder in folder_list: 
    folder_ = os.path.join(folder_path,folder) 
    FOLDER = folder[:folder.find("_Cycle")] 
    for file in tif_list:
        if FOLDER in file: #if the name the folder is in the image file name, move the image to that folder
            start = os.path.join(tif_path,file)
            end = os.path.join(folder_,file)
            os.rename(start,end)

print("Done")
```
